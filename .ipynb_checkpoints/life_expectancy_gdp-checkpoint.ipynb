```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load data
data = pd.read_csv('all_data.csv')
print(data.head(5))
```

```python
# explore data
fig, ax = plt.subplots()
sns.histplot(data=data, x='Life expectancy at birth (years)')
fig, ax = plt.subplots()
sns.histplot(data=data, x='GDP')
fig, ax = plt.subplots()
sns.lineplot(data=data, x='Year', y='Life expectancy at birth (years)', hue='Country')
fig, ax = plt.subplots()
sns.lineplot(data=data, x='Year', y='GDP', hue='Country')
fig, ax = plt.subplots()
sns.scatterplot(data=data, x='GDP',y='Life expectancy at birth (years)', hue='Country')
print(f'The average GDP is {data.GDP.mean()}, with a maximum of {data.GDP.max()} and a minimum of of {data.GDP.min()}')
mean = data['Life expectancy at birth (years)'].mean()
max = data['Life expectancy at birth (years)'].max()
min = data['Life expectancy at birth (years)'].min()
print(f'The average life expectancy is {mean}, with a maximum of {max} and a minimum of of {min}')
print('The larger the GDP, the larger the life expectancy.')
print("China's and USA's GDP have grown in the last years, while the rest of the countries is stable.")
print("Zimbabwe had a very large increase in life expectancy in the last years, which is not related to an increase in their GDP")
```
