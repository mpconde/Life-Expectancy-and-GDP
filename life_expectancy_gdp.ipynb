```python
#import libraries
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
#Load data
data = pd.read_csv('all_data.csv')
print(data.head(5))
```

```python
# explore life expectancy
sns.histplot(data=data, x='Life expectancy at birth (years)', stat="density")
mean = data['Life expectancy at birth (years)'].mean()
max = data['Life expectancy at birth (years)'].max()
min = data['Life expectancy at birth (years)'].min()
print(f'The average life expectancy is {mean}, with a maximum of {max} and a minimum of of {min}')
```

```python
# explore GDP
sns.histplot(data=data, x='GDP',stat="density")
print(f'The average GDP is {data.GDP.mean()}, with a maximum of {data.GDP.max()} and a minimum of of {data.GDP.min()}')
```

```python
# analyse life expectancy over the years
sns.lineplot(data=data, x='Year', y='Life expectancy at birth (years)', hue='Country')
print("Zimbabwe had a very large increase in life expectancy in the last years, while the rest of the countries shown a very smooth growth")
```

```python
#analyse GDP over the years
sns.lineplot(data=data, x='Year', y='GDP', hue='Country')
print("China's and USA's GDP have grown in the last years, while the rest of the countries was very stable.")
```

```python
sns.scatterplot(data=data, x='GDP',y='Life expectancy at birth (years)', hue='Country')

print('Countries with large GDP, have large life expectancy. Yet there is not always a correlation.')
print("For instance, Chile has a small GDP, similar to the Zimbabwe's, but has a large life expectancy. Whereas Zimbabwe has a very small file expectancy")
print("Zimbabwe's large increase in life expectancy in the last years is not related to an increase in their GDP")
```
